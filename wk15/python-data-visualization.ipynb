```python
# cumsum takes each element and sums of 1st and 2nd and so on from origional value

dataset = randn(50).cumsum()

dataset
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

#specify and plot dataset with black dashed line
plt.plot(dataset, 'k--')
```

```python
data = randn(30).cumsum()
data
```

```python
df1 = pd.DataFrame(data)
df1
```

```python
plt.plot(data)
```

```python
plt.plot(data, drawstyle='steps-post')
```

```python
# plot 2 lines in same plot

plt.plot(data, 'k--')
plt.plot(data, drawstyle='steps-post')
```

```python
# using diplsay labels

plt.plot(data, 'k--', label='Dashed-Line')
plt.plot(data, drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

```python
# labels in pandas

d = {'col1': [1,2,4,8,13,9,4,8,8,5], 'col2': [5,8,9,6,5,3,2,10,13,16]}

df2 = pd.DataFrame(d)
df2
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(df2['col1'],'k.',label='col1') #k. is also line style
ax.plot(df2['col2'],'k--',label='col2')
ax.legend(loc='best')
```
